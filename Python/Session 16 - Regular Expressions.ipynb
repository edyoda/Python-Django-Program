```python
s = "ababaacdfegabafdfaab"
r = re.compile("ab")
l = re.findall(r,s)
print(l)
```

```python
Regx Metachars :
    . => match with all the chars(alpha,digits,specialchars) => single occurance
    [A-Za-z] => match with only alpha => single occurance a or b or c or d .....or z
    [0-9] = > digit class => single occurance  [0-100] => not allowed 
    + => atleast once occurance 
    * => zero or more 
    ^ => start of the string 
    $ => end of the string 
    [a-z]{5}
    [0-9]{2,5}
    ? : optional 
```

```python
s = "1234ABCDE1234A564"
# 5 chars 
# 4 alpha 
# 1 char
r = re.compile("^[A-Z]{5}[0-9]{4}[A-Z]$")
m = re.search(r,s)
if m:
    print(m.group())
else:
    print("Invalid pattern")
```

```python
s = "abcd12@xyz.co"
r = re.compile("^[a-z0-9]+@[a-z0-9]+(\.[a-z]{2,3}){1,2}$")
m = re.search(r,s)
if m:
    print(m.group())
else:
    print("Invalid email")
```

```python
s = "31-55-2019"
# dd-mm-yyyy
r =  re.compile("(?P<day>{2})-(?P<month>[0-9]{2})-(?P<year>[0-9]{4})")
s =  re.search(r,s)
if s:
    print(s.group('day'))
else:
    print("Invalid date")
```

```python
s = "(+91)6123456789"
# s1 = "6123456789"
r = re.compile("^(\(\+91\))?([6-9][0-9]{9})$")
m = re.search(r,s)

if m:
    print(m.group(2))
else:
    print("Invalid contact number")
```

```python
s = "(+91)6123456789"
# s1 = "6123456789"
r = re.compile("^(?P<c_code>\(\+91\))?(?P<number>[6-9][0-9]{9})$")
m = re.search(r,s)

if m:
    print(m.group('number'))
else:
    print("Invalid contact number")
```
